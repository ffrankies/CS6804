```python
import pandas as pd
import numpy as np
```

```python
dataset = "citeseer"
```

```python
df = pd.read_csv("{0}/{0}.content".format(dataset), sep='\t', header=None)
```

```python
# Get dict of vertex ID -> 0-N ID
vertices = dict()
for index, vertex in enumerate(df[0]):
    vertices[str(vertex)] = index + 1
df['Vertex ID'] = df[0].astype(str).map(vertices)
df = df.sort_values('Vertex ID')
df.loc[df[0] == "kohrs99improving"]
```

```python
# Get dict of topic -> community ID
topics = dict()
topics_column = df.columns.size - 2  # Vertex ID is now df.columns.size - 1
for index, topic in enumerate(np.unique(df[topics_column])):
    topics[topic] = index + 1
df['Topic'] = df[topics_column].map(topics)
df
```

```python
df[['Vertex ID', 'Topic']].to_csv("{0}/{0}_truePartition.tsv".format(dataset), header=False, index=False, sep='\t')
```

```python
graph_df = pd.read_csv("{0}/{0}.cites".format(dataset), header=None, sep='\t')
graph_df["To"] = graph_df[0].astype(str).map(vertices)
graph_df["From"] = graph_df[1].astype(str).map(vertices)
for index, row in graph_df.iterrows():
    if row[0] not in vertices:
        print(row)
        
print(vertices["tobies99pspace"])
```
